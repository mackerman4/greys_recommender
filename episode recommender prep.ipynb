### Read from SQL

```python
episodes_df = pd.read_sql('''SELECT * FROM episodes_table''', conn)
episodes_df.head()
```

### Create unique season/episode ID

```python
def create_ep_id(season, ep_num):
    s_season = str(season)
    s_ep = str(ep_num).rjust(2,'0')
    return s_season+s_ep
```

```python
episodes_df['ep_id'] = episodes_df.apply(lambda x: str(x.season) + str(x.ep_num).rjust(2,'0'), axis=1)
```

```python
episodes_df.head()
```

### Calculate cosine similarities

```python
# Vectors
vectors = episodes_df.drop(columns=['ep_name', 'season', 'ep_num', 'ep_id', 'topic', 'score'])
vectors.head()
```

```python
# Calculate similarity based on cosine distance
from sklearn.metrics.pairwise import cosine_similarity

cos_sim = cosine_similarity(vectors)
cos_indices = np.vstack([np.argsort(-arr) for arr in cos_sim])
```

```python
# Find most similar eps for each case
sim_eps = []
for i in range(len(cos_indices)):
    sim_eps.append(cos_indices[i][1:11])
len(sim_eps)
```

### Create dict of episodes with name, season/ep id, top topic, similar eps

```python
ep_dict = {}
sim_eps_dict = {}
for i in range(len(sim_eps)):
    temp = {}
    for x in range(10):
        temp.update({x+1:{'name':episodes_df['ep_name'][sim_eps[i][x]], 
                                    'id':episodes_df['ep_id'][sim_eps[i][x]]}})
    sim_eps_dict.update({i:temp})
    
    ep_dict.update({i:{'name':episodes_df['ep_name'][i], 'id':episodes_df['ep_id'][i], 
                      'topic':episodes_df['topic'][i], 'sim_eps':sim_eps_dict[i]}})
```

```python
ep_dict
```

### Make ep_id the outer key for each episode

```python
id_list = []
for i in range(len(ep_dict)):
    id_list.append(ep_dict[i]['id'])
len(id_list)
```

```python
ep_id_dict = {}
for i in range(len(ep_dict)):
    del ep_dict[i]['id']
    ep_id_dict.update({id_list[i]:ep_dict[i]})   
ep_id_dict
```
