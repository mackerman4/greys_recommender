```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ep_topic_df = pd.read_csv('episodes_by_topic.csv', index_col=0)
ep_topic_df.head()
```

```python
season_topic_df = ep_topic_df.groupby(['season','topic']).size().reset_index()
season_topic_df.head()
```

```python
season_topic_df.rename(columns={'0':'count'}, inplace=True)
season_topic_df.head()
```

```python
topic_by_season = season_topic_df.pivot(index='season', columns='topic').fillna(value=0).reset_index()
topic_by_season.head()
```

```python
topic_by_season.columns = topic_by_season.columns.get_level_values(1)
```

```python
topic_by_season.head()
```

```python
topic_by_season.rename(columns={"":"season"}, inplace=True)
topic_by_season.head()
```

```python
topic_by_season.to_csv('topic_by_season.csv')
```
